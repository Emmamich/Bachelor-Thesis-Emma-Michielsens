```python
import pandas as pd
df = pd.read_excel('/content/results_alle_tweets_topic.xlsx')
```

```python
from transformers import pipeline
model_path = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
sentiment_classifier = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path)
```

```python
sentiments = []
for tweet in df['text']:
    result = sentiment_classifier(tweet)
    sentiments.append((tweet, result[0]['label'], result[0]['score']))
```

```python
sentiment_df = pd.DataFrame(sentiments, columns=['Text', 'Sentiment', 'Certainty score'])
print(sentiment_df.head())
```

```python
df['Sentiment'] = sentiment_df['Sentiment']
df['Certainty score'] = sentiment_df['Certainty score']
```

```python
output_filename = 'results_alle_tweets_sentiment.xlsx'
df.to_excel(output_filename, index=False)
```
